```python
#import the relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import pi

# Step 1: Create a sample dataset consisting of lists in a dictionary and Performance Scores of the Products as Rating
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product D'],
    'Quality': [8, 7, 6, 9],
    'Price': [5, 6, 7, 6],
    'Usability': [7, 8, 9, 7],
    'Features': [8, 6, 8, 7],
    'Customer Support': [6, 8, 7, 9]
}

# Convert the data in dictionary into a DataFrame using the instantiated object of pandas library
df = pd.DataFrame(data)

# Step 2: Set the number of categories using a list of strings for the products defined in the above data dictionary
categories = ['Quality', 'Price', 'Usability', 'Features', 'Customer Support']

# Step 3: Normalize the data (this step ensures all values are on the same scale)
# In this case, we'll assume that the ratings are on a scale of 1-10, so no further scaling is needed.

# Step 4: Prepare the data for plotting taking the number of categories we have defined in the categories list as number of axes for the chart
num_vars = len(categories)

# Angle for each axis (dividing the circle into equal sections)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# To complete the circle, append the first value to the end of the list
angles += angles[:1]

# Step 5: Create radar chart for each product
fig, ax = plt.subplots(figsize=(6, 6), dpi=100, subplot_kw=dict(polar=True))

# Plot each product
for i, product in enumerate(df['Product']):
    values = df.iloc[i, 1:].tolist()  # Extract the product values
    values += values[:1]  # Close the loop

    ax.plot(angles, values, label=product, linewidth=2, linestyle='solid')
    ax.fill(angles, values, alpha=0.25)  # Optional: fill the area under the line

# Step 6: Customize the chart
ax.set_yticklabels([])  # Remove radial ticks
ax.set_xticks(angles[:-1])  # Set the angle labels for categories
ax.set_xticklabels(categories, fontsize=12, color='black')

# Step 7: Add a title
plt.title('Product Comparison - Radar Chart', size=15, color='black', fontweight='bold', y=1.1)

# Step 8: Display legend
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))

# Show the plot
plt.show()
```
